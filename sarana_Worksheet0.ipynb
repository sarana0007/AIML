```python
def convert_units():
    """
    Converts units of length, weight, and volume.

    Prompts the user to select conversion type and value,
    performs the conversion, and prints the result.

    Returns:
        None
    """
    try:
        print("Choose conversion type: length, weight, volume")
        conversion_type = input("Enter conversion type: ").lower()

        value = float(input("Enter value to convert: "))

        if conversion_type == "length":
            print("Options: m_to_ft or ft_to_m")
            option = input("Choose option: ")

            if option == "m_to_ft":
                result = value * 3.28084
                print("Result:", result, "ft")

            elif option == "ft_to_m":
                result = value * 0.3048
                print("Result:", result, "m")

            else:
                print("Invalid length conversion")

        elif conversion_type == "weight":
            print("Options: kg_to_lbs or lbs_to_kg")
            option = input("Choose option: ")

            if option == "kg_to_lbs":
                result = value * 2.20462
                print("Result:", result, "lbs")

            elif option == "lbs_to_kg":
                result = value * 0.453592
                print("Result:", result, "kg")

            else:
                print("Invalid weight conversion")

        elif conversion_type == "volume":
            print("Options: l_to_gal or gal_to_l")
            option = input("Choose option: ")

            if option == "l_to_gal":
                result = value * 0.264172
                print("Result:", result, "gal")

            elif option == "gal_to_l":
                result = value * 3.78541
                print("Result:", result, "L")

            else:
                print("Invalid volume conversion")

        else:
            print("Unsupported conversion type")

    except ValueError:
        print("Invalid input! Please enter numeric value.")


convert_units()
```

```python
def calculate_sum(numbers):
    """Returns the sum of the numbers."""
    return sum(numbers)

def calculate_average(numbers):
    """Returns the average of the numbers."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns the maximum number."""
    return max(numbers)

def find_min(numbers):
    """Returns the minimum number."""
    return min(numbers)


try:
    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    if len(nums) == 0:
        raise ValueError("List cannot be empty")

    operation = input("Choose operation (sum, avg, max, min): ")

    if operation == "sum":
        print("Sum:", calculate_sum(nums))

    elif operation == "avg":
        print("Average:", calculate_average(nums))

    elif operation == "max":
        print("Maximum:", find_max(nums))

    elif operation == "min":
        print("Minimum:", find_min(nums))

    else:
        print("Invalid operation")

except ValueError:
    print("Invalid input! Please enter numbers.")
```

```python
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other([1,2,3,4,5,6]))
```

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]

print(get_sublist([1,2,3,4,5,6],2,4))
```

```python
def reverse_list(lst):
    return lst[::-1]

print(reverse_list([1,2,3,4,5]))
```

4.3 Exercise on Nested List

1 Flatten Nested List

```python
def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(i)
        else:
            result.append(i)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

2 Access Nested Element

```python
def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

3 Sum Nested List

```python
def sum_nested(lst):
    total = 0
    for i in lst:
        if isinstance(i,list):
            total += sum_nested(i)
        else:
            total += i
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

4 Remove Element from Nested List

```python
def remove_element(lst, elem):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.append([x for x in i if x != elem])
        elif i != elem:
            result.append(i)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

5 Find Maximum in Nested List

```python
def find_max(lst):
    max_val = float('-inf')
    for i in lst:
        if isinstance(i,list):
            max_val = max(max_val, find_max(i))
        else:
            max_val = max(max_val, i)
    return max_val

print(find_max([[1,2],[3,[4,5]],6]))
```

6 Count Occurrences

```python
def count_occurrences(lst, elem):
    count = 0
    for i in lst:
        if isinstance(i,list):
            count += count_occurrences(i,elem)
        elif i == elem:
            count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

7 Deep Flatten

```python
def deep_flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

8 Nested Average

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

Matrix Multiplication

Python Lists

```python
import time
import random

size = 100
A = [[random.random() for _ in range(size)] for _ in range(size)]
B = [[random.random() for _ in range(size)] for _ in range(size)]
result = [[0]*size for _ in range(size)]
start = time.time()
for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += A[i][k] * B[k][j]
end = time.time()
print("Python matrix multiplication time:", end - start)
```
